```python
india_weather = pd.DataFrame({
    "city":["mumbai","delhi","benglore"],
    "temperature":[32,45,30],
    "humidity":[80,60,78]
})
india_weather
```

```python
usa_weather = pd.DataFrame({
    "city":["newyork","chicago","orlando"],
    "temperature":[32,45,30],
    "humidity":[80,60,78]
})
usa_weather
```

```python
df = pd.concat([india_weather,usa_weather])
df
```

```python
df = pd.concat([india_weather,usa_weather],ignore_index=True)
df
```

```python
df = pd.concat([india_weather,usa_weather],keys=["India","USA"])
df
```

```python
#Merge
```

```python
df1 = pd.DataFrame({
    "city":["newyork","chicago","orlando"],
    "temperature":[21,14,35]
})
df1
```

```python
df2 = pd.DataFrame({
    "city":["chicago","newyork","orlando"],
    "humidity":[80,60,78]
})
df2
```

```python
df3 = pd.merge(df1,df2,on="city")
df3
```

```python
df1 = pd.DataFrame({
    "city":["newyork","chicago","orlando","baltimore"],
    "temperature":[21,14,35,38]
})
df1
```

```python
df2 = pd.DataFrame({
    "city":["chicago","newyork","san diego"],
    "humidity":[80,60,78]
})
df2
```

```python
df3 = pd.merge(df1,df2,on="city",how="inner")
df3
```

```python
df3 = pd.merge(df1,df2,on="city",how="outer")
df3
```

```python
df3 = pd.merge(df1,df2,on="city",how="left")
df3
```

```python
df3 = pd.merge(df1,df2,on="city",how="right")
df3
```
