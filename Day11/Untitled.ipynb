```python
# part 5 histogram
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
blood_sugar = [113,85,90,150,149,88,94,115,80,77,82,129]
plt.hist(blood_sugar,rwidth=0.8) #default bins =10
```

```python
plt.hist(blood_sugar,rwidth=0.5,bins=4)
```

```python
plt.xlabel("sugar_level")
plt.ylabel("Number of patients")
plt.title("Blood sugar chart")
plt.hist(blood_sugar,rwidth=0.95,bins=[80,100,125,150],color='g')
```

```python
blood_sugar_men = [113,85,90,150,149,88,94,115,80,77,82,129]
blood_sugar_women = [67,98,89,120,133,150,84,69,79,120,112,100]
plt.xlabel("sugar_level")
plt.ylabel("Number of patients")
plt.title("Blood sugar chart")
plt.hist([blood_sugar_men,blood_sugar_women],rwidth=0.95,bins=[80,100,125,150],color=['green','orange'],label=['men','women'])
plt.legend()
```

```python
exp_value = [1400,600,300,425,250]
exp_labels = ['Home Rent','Food','Internet Bill','Car','Other Utilities']
plt.pie(exp_value,labels=exp_labels)
```

```python
plt.pie(exp_value,labels=exp_labels,shadow=True)
plt.axis('equal')
plt.show()
```

```python
plt.pie(exp_value,labels=exp_labels,shadow=True,autopct="%1.1f%%",radius=1.5)
plt.axis('equal')
plt.show()
```

```python
plt.pie(exp_value,labels=exp_labels,shadow=True,autopct="%1.1f%%",radius=1.5,explode=[0,0,0,0.1,0.2])
plt.axis('equal')
plt.show()
```

```python
plt.pie(exp_value,labels=exp_labels,shadow=True,autopct="%1.1f%%",radius=1.5,explode=[0,0,0,0.1,0.2],counterclock=True,startangle=45)
plt.axis('equal')
plt.show()
```

```python
plt.pie(exp_value,labels=exp_labels,shadow=True,autopct="%1.1f%%",radius=1.5,explode=[0,0,0,0.1,0.2],counterclock=True,startangle=45)
plt.axis('equal')
plt.legend()
plt.legend(loc='upper right',shadow=True,fontsize='large')
plt.show()
```
