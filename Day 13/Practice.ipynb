```python
#  view
a = array([101,0,103,104,105])
b = a.view()
a[1] = 80
b[2] = 300
print('a = ',a)
print('b = ',b)
print('Location of a = ',id(a))
print('Location of b = ',id(b))
      
```

```python
a = array([101,200,103,104,105])
b = copy(a)
a[1] = 80
b[2] = 300
print('a = ',a)
print('b = ',b)
```
