- Every decision rules has some conditions.
- The dipper the tree more complex the decision rule and fitter the model.

# Advantages of Decision Tree 

- A decision tree does not require scaling of data as well.
- Decision tree does not require normalization of data.

# Disadvantages of Decision Tree

- A small change in data can cause a large change in the structure of the decision tree causing instability.

# Implementation of Decision Tree Algorithm

# Import the necessary libraries

```python
import numpy as np
import pandas as pd
```

# Import/Load the dataset

```python
dataset = pd.read_csv(R'/kaggle/input/diabetes-dataset/diabetes.csv')
dataset
```

```python
from sklearn.preprocessing import LabelEncoder
l1=LabelEncoder()
dataset["outcome"]=l1.fit_transform(dataset["outcome"])
```

# Independent Variable (X) & Dependent Variable (Y)

```python
X = dataset.iloc[:, :-1]
Y = dataset.iloc[:, -1]
```

# Split the dataset for train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

# Model training

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, Y_train)
```

# Model prediction

```python
y_pred =classifier.predict(X_test)
print(y_pred)
```

# Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, y_pred)
print(cm)
```

# Making the Accuracy Score

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(Y_test, y_pred)*100
ac
```
