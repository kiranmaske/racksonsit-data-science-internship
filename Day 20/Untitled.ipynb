```python
import cv2

save_image = 5150
cap = cv2.VideoCapture(r"D:\Internship_June2025\4th_Week\open cv\4K Road traffic video.mp4")
# variables
sec = 0.0
framerate = 1

try:
    # Prompt user for number of frames to capture
    n = int(input("Enter the count: "))
    
    for x in range(n):
        ret, img = cap.read()

        # Check if the frame was read successfully
        if not ret:
            print("Unable to read frame. Exiting loop.")
            break

        print(f"Image capture {save_image}")
        filename = f"D:\\Internship_June2025\\4th_Week\\Day2\\img_{save_image}.jpg"
        
        # Save the frame as an image file
        cv2.imwrite(filename, img)
        save_image += 1

        # Skip frames based on framerate
        for _ in range(framerate):
            cap.read()

except Exception as e:
    print(f"There is an exception - {e}")

# Release the video capture object
cap.release()
```
