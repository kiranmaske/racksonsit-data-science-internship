```python
## import libraries ##

import cv2

## Cascade selection ##

face_cascade = cv2.CascadeClassifier(r"D:\Internship_June2025\4th_Week\open cv\haarcascade_eye.xml")
face_cascade

## Read Original Image ##

img = cv2.imread(r"D:\Internship_June2025\4th_Week\open cv\Kat.jpg")

## Convert original image to gray ##

gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

## detect multiple images ##

faces = face_cascade.detectMultiScale(gray,1.1,2)

## draw an rectangle around face ##

for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
cv2.imshow('This is katrina kaif',img)
cv2.waitKey()
```
