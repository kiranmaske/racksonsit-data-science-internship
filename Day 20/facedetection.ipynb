```python
import cv2
face_cascade=cv2.CascadeClassifier(r'E:\keras project\haarcascade_eye.xml')
webcam = cv2.VideoCapture(0)
while True:
    _,img=webcam.read()
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(gray,1.1,2)
    for(x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(201,0,118),5)
    cv2.imshow("Face Detection",img)
    key=-cv2.waitKey(10)
    if key==27:
        break
webcam.release()
cv2.destroyAllWindows()
```
