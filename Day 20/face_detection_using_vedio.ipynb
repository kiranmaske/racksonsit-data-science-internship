```python
import cv2

face_cascade=cv2.CascadeClassifier(r"D:\Internship_June2025\4th_Week\open cv\haarcascade_eye.xml")

cap=cv2.VideoCapture(0)

_,img=cap.read()

gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

faces=face_cascade.detectMultiScale(gray,1.1,2)

for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(201,0,118),5)
    cv2.imshow('This is kiran',img)
    k=cv2.waitKey(1) & 0xff
    if k==27:
        break

cap.release()
```
