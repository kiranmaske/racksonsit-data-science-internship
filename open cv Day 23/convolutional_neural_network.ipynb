# Convolutional Neural Network

### Importing the libraries

```python
import tensorflow as tf
from keras.src.legacy.preprocessing.image import ImageDataGenerator
```

```python
tf.__version__
```

```python
# rescale = 1./255:

# Scales the pixel values from 0–255 to 0–1.

# Normalization helps the model converge faster.

# shear_range = 0.2:

# Applies shearing transformation (slanting the image), like skewing horizontally/vertically by 20%.

# zoom_range = 0.2:

# Randomly zooms in on images by up to 20%.

# horizontal_flip = True:

# Randomly flips the image horizontally (e.g., a cat facing left becomes a cat facing right)
```

## Part 1 - Data Preprocessing

### Preprocessing the Training set

```python
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
training_set = train_datagen.flow_from_directory(r'D:\Data Science All\Data Science AI ML 45 Days Training\DL_PPTS\dogs-vs-cats\train',
                                                 target_size = (64, 64),
                                                 batch_size = 32,
                                                 class_mode = 'binary')
```

### Preprocessing the Test set

```python
test_datagen = ImageDataGenerator(rescale = 1./255)
test_set = test_datagen.flow_from_directory(r'D:\Data Science All\Data Science AI ML 45 Days Training\DL_PPTS\dogs-vs-cats\test1',
                                            target_size = (64, 64),
                                            batch_size = 32,
                                            class_mode = 'binary')
```

## Part 2 - Building the CNN

### Initialising the CNN

```python
cnn = tf.keras.models.Sequential()
```

### Step 1 - Convolution

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=[64, 64, 3]))
```

```python
# filters=32: Creates 32 convolution filters (also called feature detectors).

# kernel_size=3: Each filter is 3x3 in size.

# activation='relu': Applies ReLU (Rectified Linear Unit) to introduce non-linearity.

# input_shape=[64, 64, 3]: Input image size — 64x64 pixels with 3 color channels (RGB).
```

### Step 2 - Pooling

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
# pool_size=2: Looks at 2x2 windows in the feature map.

# strides=2: Moves 2 steps at a time (no overlap).

# Effect: Reduces spatial size (downsampling), keeps only strongest features → improves efficiency and reduces overfitting.
```

### Adding a second convolutional layer

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
# Another 32 filters (3x3) applied to previously pooled features.

# Learns more complex patterns like shapes, curves, eyes, etc.
```

### Step 3 - Flattening

```python
cnn.add(tf.keras.layers.Flatten())
```

```python
# Converts the 2D feature maps into a 1D feature vector to feed into Dense layers.

# For example, a (16x16x32) output becomes a vector of 8192 values.
```

### Step 4 - Full Connection

```python
cnn.add(tf.keras.layers.Dense(units=128, activation='relu'))
```

```python
# 128 neurons.

# ReLU activation.

# Processes flattened features and learns complex decision boundaries.
```

### Step 5 - Output Layer

```python
cnn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
# 1 neuron because it's a binary classification (cat vs dog).

# Sigmoid activation gives a probability between 0 and 1.

# Output close to 0 → class 0 (dog)

# Output close to 1 → class 1 (cat)
```

## Part 3 - Training the CNN

### Compiling the CNN

```python
cnn.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

### Training the CNN on the Training set and evaluating it on the Test set

```python
cnn.fit(x = training_set, validation_data = test_set, epochs = 25)
```

## Part 4 - Making a single prediction

```python
import numpy as np
from keras.preprocessing import image
test_image = image.load_img('D:\Data Science All\Data Science AI ML 45 Days Training\DL_PPTS\dogs-vs-cats\single_prediction\cat_or_dog_1.jpg',
                            target_size = (64, 64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis = 0)
result = cnn.predict(test_image)
training_set.class_indices
if result[0][0] == 0:
  prediction = 'dog'
else:
  prediction = 'cat'
```
